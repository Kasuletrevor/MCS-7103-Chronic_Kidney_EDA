```python
# Selecting only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['float64', 'int64'])
```

```python
correlation_matrix = numeric_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Numeric Variables')
plt.show()
```

```python
# Extract pairs of variables with correlation above 0.7 or below -0.7
high_correlation_pairs = []

# Iterating through the correlation matrix to find high correlations
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        correlation_value = correlation_matrix.iloc[i, j]
        if abs(correlation_value) > 0.7:  # Correlation threshold
            high_correlation_pairs.append((
                correlation_matrix.columns[i],  # Variable 1
                correlation_matrix.columns[j],  # Variable 2
                correlation_value  # Correlation value
            ))

# Displaying high correlation pairs
high_correlation_pairs
```

```python
# Imputation for numerical variables using median
data['age'].fillna(data['age'].median(), inplace=True)
data['bp'].fillna(data['bp'].median(), inplace=True)
data['sg'].fillna(data['sg'].median(), inplace=True)
data['al'].fillna(data['al'].median(), inplace=True)
data['su'].fillna(data['su'].median(), inplace=True)
data['bgr'].fillna(data['bgr'].median(), inplace=True)
data['bu'].fillna(data['bu'].median(), inplace=True)
data['sc'].fillna(data['sc'].median(), inplace=True)
data['sod'].fillna(data['sod'].median(), inplace=True)
data['pot'].fillna(data['pot'].median(), inplace=True)
data['hemo'].fillna(data['hemo'].median(), inplace=True)
data['pcv'].fillna(data['pcv'].median(), inplace=True)
data['wc'].fillna(data['wc'].median(), inplace=True)
data['rc'].fillna(data['rc'].median(), inplace=True)
```

```python

# Imputation for categorical variables using the most frequent value (mode)
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']
for column in categorical_columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
# Checking again for any remaining missing values
missing_values_after_imputation = data.isnull().sum()

missing_values_after_imputation
```

```python
# Classify columns into categorical and numerical in the current data
cat_cols = [col for col in data.columns if data[col].dtype == 'object']
num_cols = [col for col in data.columns if data[col].dtype != 'object']
```

```python
# Display unique values in each categorical column to identify non-standard values
non_standard_values = {}
for col in cat_cols:
    unique_values = data[col].unique()
    non_standard_values[col] = unique_values
    print(f'{col} has {unique_values} values\n')

# Displaying the non-standard values to the user for review
non_standard_values
```

```python
# View the specific values in the categorical columns
for col in cat_cols:
    print(f'{col} has {data[col].unique()} values\n')
```

```python
# Cleaning non-standard values in 'dm' and 'cad' columns
data['dm'] = data['dm'].str.strip().replace({'\tyes': 'yes', '\tno': 'no', ' yes': 'yes'})
data['cad'] = data['cad'].str.strip().replace({'\tno': 'no'})

# Verifying the cleaned values
cleaned_values = {
    'dm': data['dm'].unique(),
    'cad': data['cad'].unique()
}

cleaned_values
```
